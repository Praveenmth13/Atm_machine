```python
import time
print("Welcome to HDFC Bank ")
time.sleep(2)

pin = 4660
attempt = 3
balance = 50000
while attempt != 0:
    user_pin = int(input("Please enter your PIN*"))
    if user_pin != pin:
        attempt -= 1
        print("Invalid PIN*")
        print(f"You have {attempt} number of attempt left")
        
    else:
        print("To select your transaction")
        user_choice = input("B : Balance Enquiry , D : Cash Deposite, W : Cash Withdrawl")
        
        
        if user_choice == "B":
            print(f"Your total balance is Rs.{balance}")
            
            
        if user_choice == "D":
            deposit_user = int(input("Enter the amount :"))
            total_balance = deposit_user + balance
            print(f"You have deposited Rs.{deposit_user}")
            print(f"You total balance is Rs.{total_balance}")
            
            
        if user_choice == "W":
            withdraw_user = int(input("Enter amount to be withdrawn:"))
            total_balance = balance - withdraw_user
            print(f"You have withdrawn Rs.{total_balance}")
            print(f"You total balance is Rs.{total_balance}")
            
    user_exit = input("Would you like to exit? Yes/No")
    if user_exit == "Yes":
        print("Thank You for banking with us")
        time.sleep(2)
        break
    else:
        continue
        
```
